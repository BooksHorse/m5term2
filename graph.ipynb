## นำเข้า library
- pandas ใช้อ่านข้อมูล .csv , สร้างแผนภูมิแท่ง (bar)
- matplotlib ใช้สร้างแผนภูมิ กล่อง,กระจาย (box,scatter)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## นำเข้าข้อมูลจากไฟล์ `average-income.csv` และแสดง

```python
df_expense=pd.read_csv("average-income.csv") # นำเข้าข้อมูลจากไฟล์ `average-income.csv`
df_expense # แสดง df_expense
```

## แสดงตารางแบบ Transpose (จากแนวนอนเป็นแนวตั้ง ▬ -> ▮)

```python
df_expense=df_expense.set_index("Region and province") # เลือกหัวตารางที่ชื่อว่า `Region and province`
```

```python
y=df_expense.T # Transpose ตาราง (จากแนวตั้งเป็นแนวนอน)
y # แสดง y
```

## สร้างกราฟเส้น

```python
plt.plot(y,marker='P') # plot ตาราง
plt.xlabel("Year",fontsize=10,color="red") # ตั้งแกน x ให้บอก Year
plt.ylabel("Income",fontsize=10,color="green") # ตั้งแกน y ให้บอก Income
plt.grid() # ใส่ตารางในกราฟ
plt.legend(df_expense.index) # ใส่กล่องข้อความที่บอกความหมายของเส้น
```

## สร้างแผนภูมิแท่ง

```python
df_expense.plot(kind='bar',subplots=True,figsize=(5,30)) # สร้างแผนภูมิแท่ง
```

## สร้างแผนภูมิกล่อง

```python
plt.boxplot(y) # สร้างแผนภูมิกล่อง
plt.xticks(ticks=[1,2,3,4,5],labels=df_expense.index,rotation='vertical') # ใส่ข้อความด้านล่าง 5 ข้อความ Greater_Bangkok, Central_Region, ...
```
